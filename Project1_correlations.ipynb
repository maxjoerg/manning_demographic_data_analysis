Lets build up another prototype

```python
formula = "LifeExp2018 ~ Education_No_Highschool_pct + Region + income_mean + \
    Education_Bachelor_only_pct  + Education_No_Highschool_pct : Region + income_mean : Region + \
    Education_Bachelor_only_pct : Region "
results = smf.ols(formula, data=df_final).fit()
print(results.summary())
```

ok income with region brought no benefit, Bachelor only as well

```python
formula = "LifeExp2018 ~ Education_No_Highschool_pct + Region + income_mean + \
    Education_No_Highschool_pct : Region "
    
results = smf.ols(formula, data=df_final).fit()
print(results.summary())
```

we can identify a problem with income_mean:  a very tiny coeeficient due to the fact, that the data is not scaled. So let us scale income_mean.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_income_mean =  scaler.fit_transform( df_final.income_mean.values.reshape(-1, 1))
df_final["scaled_income_mean"]  = scaled_income_mean
df_final.head()
```

```python
### let us try again
formula = "LifeExp2018 ~ Education_No_Highschool_pct + Region + scaled_income_mean + \
    Education_No_Highschool_pct : Region "
    
results = smf.ols(formula, data=df_final).fit()
print(results.summary())
```
